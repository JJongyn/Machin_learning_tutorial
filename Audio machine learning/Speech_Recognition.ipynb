```python
import speech_recognition as sr

r = sr.Recognizer()
```

### 다양한 기업에서 제공하는 음성 인식 기술을 사용 가능
* `recognize_google()`
* `recognize_google_cloud()`
* `recognize_bing()`
* `recognize_houndify()`
* `recognize_ibm()`

```python
american_file = '/content/drive/MyDrive/american.wav'
```

```python
import librosa.display
import IPython.display as ipd
import matplotlib.pyplot as plt

plt.style.use('seaborn-white')

fig = plt.figure(figsize=(14,4))
american_wav, rate = librosa.core.load(american_file)
librosa.display.waveplot(american_wav, sr=rate)

ipd.Audio(american_wav, rate=rate)
```

## 음성 인식 결과를 return 

```python
american_audio = sr.AudioFile('/content/drive/MyDrive/american.wav')

with american_audio as source:
  audio = r.record(source)

r.recognize_google(audio_data=audio, language='en-US')
```

`duration=2` -> 2초까지만 가져옴

```python
american_audio = sr.AudioFile('/content/drive/MyDrive/american.wav')

with american_audio as source:
  audio = r.record(source, duration=2)

r.recognize_google(audio_data=audio, language='en-US')
```

`offset=2` -> 2초부터 가져옴



```python
american_audio = sr.AudioFile('/content/drive/MyDrive/american.wav')

with american_audio as source:
  audio = r.record(source, offset=2)

r.recognize_google(audio_data=audio, language='en-US')
```
